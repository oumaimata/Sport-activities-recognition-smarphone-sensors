## Processing of the main DataFrame "data" for analysis

```python
data = pd.concat([d1, d2, d3, d4, d5, d6])
data['rowData'] =  data.index+1
data['Frequence'] = data.index%125+1

#data = data[(data["Subject"] == 1) | (data["Subject"] == 2) | (data["Subject"] == 3)]

data_left_arm = data.drop(["RA_xacc", "RA_yacc", "RA_zacc", "RA_xgyro", "RA_ygyro", "RA_zgyro", "RA_xmag", "RA_ymag", "RA_zmag"],axis=1)
data_right_arm = data.drop(["LA_xacc", "LA_yacc", "LA_zacc", "LA_xgyro", "LA_ygyro", "LA_zgyro", "LA_xmag", "LA_ymag", "LA_zmag"],axis=1)

data_left_arm['is_left'] = 1
data_right_arm['is_left'] = 0

data_left_arm.rename(columns = {"LA_xacc" : "A_xacc","LA_yacc" : "A_yacc","LA_zacc" : "A_zacc",
                                "LA_xgyro" : "A_xgyro","LA_ygyro" : "A_ygyro","LA_zgyro" : "A_zgyro",
                                "LA_xmag" : "A_xmag", "LA_ymag" : "A_ymag", "LA_zmag" : "A_zmag"}
                     ,inplace = True)
data_right_arm.rename(columns = {"RA_xacc" : "A_xacc","RA_yacc" : "A_yacc","RA_zacc" : "A_zacc",
                                "RA_xgyro" : "A_xgyro","RA_ygyro" : "A_ygyro","RA_zgyro" : "A_zgyro",
                                "RA_xmag" : "A_xmag", "RA_ymag" : "A_ymag", "RA_zmag" : "A_zmag"}
                     ,inplace = True)

data = pd.concat([data_left_arm,data_right_arm])
data.isnull().values.any() #return False
data.head()
```
